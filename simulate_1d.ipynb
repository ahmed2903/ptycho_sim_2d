```python
import numpy as np 
import matplotlib.pyplot as plt

%matplotlib ipympl
```

```python
space = np.arange(-5,5,.001)


crystal_signal = np.abs(np.sin(100*space))

T = .01                 
Fs = 100        
comb = np.zeros_like(space)
crystal_signal[::int(Fs*T)] = 1

plt.figure()
plt.plot(space, crystal_signal)

probe_signal = np.abs(np.sinc(2*space))
```

```python
plt.figure()
plt.plot(space, probe_signal)
plt.plot(space, crystal_signal)
```

```python

fft = lambda x: np.fft.fftshift(np.fft.fft(np.fft.ifftshift(x)))

exit_wave = crystal_signal * probe_signal

crystal_ft = fft(crystal_signal)
pupil = fft(probe_signal)
measured = fft(exit_wave)
```

```python
#plt.figure()
fig, ax = plt.subplots(3,1, figsize=(5,15))

ax[0].plot(crystal_ft)
ax[1].plot(pupil)
ax[2].plot(measured)


```

```python
probe2 = np.abs(np.sinc(2*space - np.pi/2))
pupil2 = fft(probe2)
exit_wave2 = probe2 * crystal_signal
```

```python
plt.figure()
plt.plot(space, crystal_signal)
plt.plot(space, probe_signal, label='first probe')
plt.plot(space, probe2, label='second probe')

```

```python
# plt.figure()
# fig, ax = plt.subplots(2,1, figsize=(5,15))
plt.figure()
plt.plot(exit_wave, label = '1')
plt.plot(exit_wave2, label='2')
plt.legend()
```

```python
plt.figure()
plt.plot(pupil, label = '1')
plt.plot(pupil2, label='2')
plt.legend()
```

```python
measured_2 = fft(exit_wave2)

# plt.figure()
# fig, ax = plt.subplots(2,1, figsize=(5,15))
plt.figure()
plt.plot(measured, label = '1')
plt.plot(measured_2, label='2')
plt.legend()
```
